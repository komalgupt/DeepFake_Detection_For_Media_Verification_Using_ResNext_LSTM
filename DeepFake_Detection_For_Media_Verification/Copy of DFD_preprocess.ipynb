```python
# Load the OpenCV DNN face detector
face_net = cv2.dnn.readNetFromCaffe(
    'deploy.prototxt',
    'res10_300x300_ssd_iter_140000.caffemodel'
)
```

```python
# Face detection function
def detect_face_dnn(image):
    h, w = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 1.0, (300, 300),
                                 (104.0, 177.0, 123.0), False, False)
    face_net.setInput(blob)
    detections = face_net.forward()

    faces = []
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > 0.9:
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (x1, y1, x2, y2) = box.astype("int")
            faces.append((y1, x2, y2, x1))  # top, right, bottom, left
    return faces
```

```python
# Extract frames from a video
def frame_extract(video_path):
    cap = cv2.VideoCapture(video_path)
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        yield frame
    cap.release()
```

```python
import glob
def create_face_videos(path_list, out_dir):
    already_present_count = glob.glob(out_dir + '*.mp4')
    print("No of videos already present ", len(already_present_count))
    for path in tqdm(path_list):
        out_path = os.path.join(out_dir, path.split('/')[-1])
        file_exists = glob.glob(out_path)
        if len(file_exists) != 0:
            print("File Already exists: ", out_path)
            continue

        out = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'), 30, (112, 112))

        for idx, frame in enumerate(frame_extract(path)):
            if idx > 150:
                break

            faces = detect_face_dnn(frame)
            if faces:
                top, right, bottom, left = faces[0]  # Use only the first detected face
                try:
                    face_crop = frame[top:bottom, left:right]
                    resized_face = cv2.resize(face_crop, (112, 112))
                    out.write(resized_face)
                except:
                    pass

        out.release()
```
